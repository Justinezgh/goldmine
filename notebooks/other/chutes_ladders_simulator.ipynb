# Test epidemiology simulator

```python
import numpy as np
import corner
from matplotlib import pyplot as plt
%matplotlib inline

from goldmine.simulators.chutes_ladders import ChutesLaddersSimulator
```

## Run simulation

```python
simulator = ChutesLaddersSimulator()
```

```python
theta1 = np.array([-0.5])
theta2 = np.array([0.])
theta3 = np.array([.5])
```

```python
x1 = simulator.rvs(theta1, 10000)
x2 = simulator.rvs(theta2, 10000)
x3 = simulator.rvs(theta3, 10000)
```

```python
plt.figure(figsize=(5,5))

plt.hist(x1, range=(-0.5,80.5), bins=81, histtype='step', label=r'$\theta=-0.5$')
plt.hist(x2, range=(-0.5,80.5), bins=81, histtype='step', label=r'$\theta=0$')
plt.hist(x3, range=(-0.5,80.5), bins=81, histtype='step', label=r'$\theta=0.5$')

plt.legend()

plt.show()
```

```python
x, t_xz = simulator.rvs_score(theta1, 1000)
```

```python
plt.figure(figsize=(5,5))

plt.hist(t_xz, bins=100, histtype='step')

plt.show()
```

```python
x, r_xz = simulator.rvs_ratio(theta2, theta1, theta3, n=1000)
```

```python
plt.figure(figsize=(5,5))

plt.hist(np.log(r_xz), bins=50, range=(-30.,30.), histtype='step')

plt.show()
```
