# Test epidemiology simulator

```python
import numpy as np
import corner
from matplotlib import pyplot as plt
%matplotlib inline

from goldmine.simulators.epidemiology import Epidemiology
from goldmine.simulators.epidemiology2d import Epidemiology2D
```

## Run simulation

```python
simulator = Epidemiology(end_time=10, delta_t=0.1, n_individuals=21)
simulator2d = Epidemiology2D(end_time=10, delta_t=0.1, n_individuals=21)
```

```python
theta = np.array([3.59,0.59,0.097,1.])
theta2d = np.array([3.59,0.097])
theta2d_alt = np.array([4,0.05])
```

```python
x, histories = simulator.rvs(theta, 1000, return_histories=True)
```

```python
x2d, histories2d = simulator2d.rvs(theta2d, 1000, return_histories=True)
```

```python
x2d_alt, histories2d_alt = simulator2d.rvs(theta2d_alt, 1000, return_histories=True)
```

## Evolution

```python
x_histories = [
    [simulator._calculate_observables(state) for state in history]
      for history in histories
]
x_histories = np.array(x_histories)

x_histories2d = [
    [simulator2d._calculate_observables(state) for state in history]
      for history in histories2d
]
x_histories2d = np.array(x_histories2d)

x_histories2d_alt = [
    [simulator2d._calculate_observables(state) for state in history]
      for history in histories2d_alt
]
x_histories2d_alt = np.array(x_histories2d_alt)
```

```python
t = np.linspace(0.,10.,len(x_histories[0,:,0]))

labels=['Shannon diversity', '# strains', 'Carriage',
        'Coinfection', 'Prevalence most common strain', '# singletons']

plt.figure(figsize=(12.,8.))

for o in range(6):
    ax = plt.subplot(2,3,o+1)
    
    for i in range(1,21):
        plt.plot(t,x_histories[i,:,o],
                 c='#CC002E', lw=0.5, alpha=0.1)
    plt.plot(t,x_histories[0,:,o],
             c='#CC002E', lw=1., alpha=0.8)
    plt.plot(t,np.mean(x_histories[:,:,o], axis=0),
             c='#CC002E', lw=2., alpha=1.,
            label='4d')
    
    plt.plot(t,np.mean(x_histories2d[:,:,o], axis=0),
             c='mediumblue', lw=2., alpha=1.,
            label='2d')
    
    plt.plot(t,np.mean(x_histories2d_alt[:,:,o], axis=0),
             c='darkgreen', lw=2., alpha=1.,
            label='2d, other theta')
        
    plt.legend()
    plt.xlabel('t')
    plt.ylabel(labels[o])
        
plt.tight_layout()
plt.show()
```
